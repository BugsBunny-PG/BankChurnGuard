```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle
from collections import Counter
```

```python
data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data.shape
data.columns
data.dtypes
counter = Counter(data['Exited'])
print(counter)
```

```python
final_data=pd.DataFrame()
final_data=data[[ 'CreditScore', 'Geography',
       'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary', 'Exited']]

final_data.head()
```

```python
en=preprocessing.LabelEncoder()

# final_data['Geography']=en.fit_transform(final_data['Geography'])
# final_data['Gender']=en.fit_transform(final_data['Gender'])
en.fit(final_data['Geography'])
final_data.loc[:, 'Geography'] = en.transform(final_data['Geography'])

en.fit(final_data['Gender'])
final_data.loc[:, 'Gender'] = en.transform(final_data['Gender'])
final_data.head()
```

```python
x_df = final_data.iloc[:, 0:-1]
y_df= final_data.drop(final_data.iloc[:, 0:-1], axis=1)
y_df.head()
```

```python
x_train, x_test, y_train, y_test =train_test_split(x_df,y_df, test_size=0.2,random_state=42)
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train.values.ravel())
```

```python
pred=model.predict(x_test)
score=accuracy_score(y_test,pred)
print(score)
```

```python
file=open('customer_churn_model.pkl', 'wb')
pickle.dump(model, file)
```
